# üóíÔ∏è Overview

Gradient Descent is one of the most fundamental algorithms in machine learning.
It‚Äôs how most models learn ‚Äî by adjusting parameters to minimize a loss (error) function.

At its core, it works like this:

1. Start with random model parameters (like weights).

2. Compute how ‚Äúwrong‚Äù the model is (loss function).

3. Calculate how much each parameter contributed to that error (gradient).

4. Update each parameter a little bit in the opposite direction of the gradient.

5. Repeat until the model stops improving.

## üß© Concept Breakdown

Key Terms:

- Loss Function ‚Üí measures how bad the model is.

- Learning Rate (Œ∑) ‚Üí controls how big a step we take each iteration.

- Gradient ‚Üí direction and magnitude of steepest ascent (we move opposite it).

## üßÆ Mathematical Idea

For a simple function 
f(x)=x¬≤,

its derivative f‚Ä≤(x)=2x tells us the slope at any point.

The update rule is:

xnew‚Äã=xold‚Äã‚àíŒ∑‚ãÖf‚Ä≤(x)

We stop when the slope is close to zero (a minimum).

## üß† Practice: Simple Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt

# Function and its derivative
def f(x):
    return x**2

def df(x):
    return 2*x

# Initialize values
x = 10       # starting point
lr = 0.1     # learning rate
epochs = 25  # iterations
history = [x]

for i in range(epochs):
    grad = df(x)
    x = x - lr * grad
    history.append(x)

# Plot progress
xs = np.linspace(-10, 10, 100)
plt.plot(xs, f(xs), label='f(x) = x^2')
plt.scatter(history, [f(h) for h in history], color='red')
plt.title('Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

print(f"Minimum at x = {x:.4f}")
```

## üîß Mini Project: Linear Regression from Scratch

Use gradient descent to fit a straight line to data ‚Äî without using scikit-learn!

### Instructions 

1. Generate synthetic data:

y = 4x + noise

2. Initialize random weights m and b (slope and intercept).

3. Define a loss function (Mean Squared Error).

4. Compute gradients w.r.t. m and b.

5. Update them using gradient descent.

6. Plot the line as it fits the data over time.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = np.linspace(0, 10, 100)
y = 4*X + np.random.randn(100)*2

# Initialize parameters
m, b = np.random.randn(), np.random.randn()
lr = 0.01
epochs = 1000

# Gradient Descent
for i in range(epochs):
    y_pred = m*X + b
    error = y_pred - y
    dm = (2/len(X)) * np.sum(error * X)
    db = (2/len(X)) * np.sum(error)
    m -= lr * dm
    b -= lr * db

print(f"Fitted line: y = {m:.2f}x + {b:.2f}")

# Plot results
plt.scatter(X, y, alpha=0.6, label='Data')
plt.plot(X, m*X + b, color='red', label='Fitted Line')
plt.legend()
plt.title('Linear Regression via Gradient Descent')
plt.show()
```

## üí¨ Reflection 

How does changing the learning rate affect convergence?

- The learning rate controls how big each update step is during training.

- If the learning rate is too high, the model may overshoot the minimum ‚Äî bouncing around or even diverging.

- If the learning rate is too low, convergence becomes very slow, taking many iterations to reach the minimum.

- The ideal learning rate finds a balance: fast but stable convergence.

What happens if you increase the noise?

- Adding more noise makes the data less predictable and more scattered.

- The model has a harder time finding a clear pattern, so the loss function becomes bumpier.

- This can lead to a less accurate fit and more variance in the model‚Äôs predictions.

- In extreme cases, it might cause the model to overfit the noise instead of learning the true relationship.

Why does gradient descent sometimes fail to converge?

- The learning rate is too large, causing the steps to jump over the minimum.

- The loss surface might be very complex (non-convex), trapping the algorithm in local minima.

- Poor feature scaling can distort the gradient directions, slowing or preventing convergence.

- Noisy or inconsistent gradients (especially with stochastic methods) can make convergence unstable.
