# üß† Day 38 ‚Äì Improving Neural Networks

Today we‚Äôll learn how to improve the performance of your neural networks using **dropout, batch normalization, and hyperparameter tuning**. These techniques help prevent overfitting and make your model generalize better.

---

## Goals
- Understand **overfitting** vs **underfitting**  
- Implement **dropout layers**  
- Apply **batch normalization**  
- Tune hyperparameters (learning rate, number of neurons, epochs)  
- Evaluate model improvements


````python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
````

````python
# Generate synthetic binary classification data
np.random.seed(42)
X = np.random.randn(500, 2)
y = (X[:, 0] * X[:, 1] > 0).astype(int)  # XOR-ish pattern
````

````python
# Build improved ANN with dropout and batch normalization
model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),
    BatchNormalization(),
    Dropout(0.3),
    Dense(4, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=Adam(learning_rate=0.01),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
````

````python
# Train the model
history = model.fit(X, y, epochs=100, batch_size=32, verbose=0)
````

````python
# Plot training loss and accuracy
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['loss'])
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'])
plt.title("Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.show()
````

## ‚úÖ Summary

- Added **dropout** to hidden layers to prevent overfitting  
- Applied **batch normalization** to stabilize training  
- Tuned **learning rate** to improve convergence  
- Observed improved generalization on synthetic data
