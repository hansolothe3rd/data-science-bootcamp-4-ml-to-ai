## üîπ Recap: ANN Components

- **Input layer** ‚Äì receives features  
- **Hidden layers** ‚Äì learn patterns  
- **Output layer** ‚Äì predicts target  

Activation functions:
- **ReLU** ‚Äì hidden layers  
- **Sigmoid** ‚Äì binary output  
- **Softmax** ‚Äì multi-class output  

---

## üìä Dataset

We‚Äôll use a synthetic dataset for binary classification with 2 inputs.

---

## üèóÔ∏è Model Architecture

- Input ‚Üí Hidden layer 1 (8 neurons, ReLU)  
- Hidden layer 2 (4 neurons, ReLU)  
- Output layer (1 neuron, Sigmoid)  

We‚Äôll train for 100 epochs with Adam optimizer.


```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Generate synthetic binary classification data
np.random.seed(42)

X = np.random.randn(500, 2)
y = (X[:, 0] * X[:, 1] > 0).astype(int)  # nonlinear pattern

```

```python
# Build the ANN
model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# Train the model
history = model.fit(X, y, epochs=100, verbose=0)
```

```python
# Plot loss and accuracy
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['loss'])
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'])
plt.title("Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.show()
```

## ‚úÖ Summary

- Built a 2-hidden-layer ANN  
- Trained on synthetic data for binary classification  
- Observed loss decreasing and accuracy increasing

## üìì Mini Project ‚Äì Build & Train Your Own ANN

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
np.random.seed(123)

# 500 samples, 2 features
X = np.random.randn(500, 2)
y = (X[:, 0] * X[:, 1] > 0).astype(int)  # XOR-style pattern
```

```python
model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),
    Dense(6, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(X, y, epochs=100, batch_size=32, verbose=0)
```

```python
plt.figure(figsize=(12,5))

# Loss
plt.subplot(1,2,1)
plt.plot(history.history['loss'])
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

# Accuracy
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'])
plt.title("Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.show()
```

- The training loss should decrease over epochs.  
- Accuracy should increase and ideally reach close to 1.0.  
- If it plateaus early, try adjusting hidden layer sizes, activations, or learning rate.  
- This gives hands-on practice building and tuning your first multi-layer neural network.
